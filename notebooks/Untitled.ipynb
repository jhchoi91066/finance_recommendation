```python
import pandas as pd
import glob
import os

# CSV 파일 경로 지정
csv_files = glob.glob(os.path.join("/Users/jinhochoi/Desktop/개발/finance_recommendation/data", "*.csv"))

# 여러 파일 불러와서 하나의 DataFrame으로 합치기
data_list = []
for file in csv_files:
    df = pd.read_csv(file)
    data_list.append(df)

data = pd.concat(data_list, ignore_index=True)

# 기본 확인
print(data.head())
print(data.info())
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# 간단히 user-item 매트릭스 만들기
# client_id = 사용자, merchant_id = 아이템, amount = 거래 금액
interaction = data.groupby(['client_id', 'merchant_id'])['amount'].count().reset_index()
interaction_matrix = interaction.pivot(index='client_id', columns='merchant_id', values='amount').fillna(0)

print("User-Item Matrix 크기:", interaction_matrix.shape)
display(interaction_matrix.head())
```

```python
# 유저간 유사도 계산
user_similarity = cosine_similarity(interaction_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=interaction_matrix.index, columns=interaction_matrix.index)

# 예시: 특정 사용자와 비슷한 사용자 찾기
target_user = interaction_matrix.index[0]
similar_users = user_similarity_df[target_user].sort_values(ascending=False)[1:6]  # 자기 자신 제외
print("Target User:", target_user)
print("Similar Users:", similar_users)
```

```python
# 아이템간 유사도 계산
item_similarity = cosine_similarity(interaction_matrix.T)
item_similarity_df = pd.DataFrame(item_similarity, index=interaction_matrix.columns, columns=interaction_matrix.columns)

# 예시: 특정 아이템과 비슷한 아이템 찾기
target_item = interaction_matrix.columns[0]
similar_items = item_similarity_df[target_item].sort_values(ascending=False)[1:6]
print("Target Item:", target_item)
print("Similar Items:", similar_items)
```
