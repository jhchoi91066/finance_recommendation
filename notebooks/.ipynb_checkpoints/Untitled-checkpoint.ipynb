```python
import pandas as pd
import glob
import os

# CSV 파일 경로 지정
csv_files = glob.glob(os.path.join("../data", "*.csv"))

# 여러 파일 불러와서 하나의 DataFrame으로 합치기
data_list = []
for file in csv_files:
    df = pd.read_csv(file)
    data_list.append(df)

data = pd.concat(data_list, ignore_index=True)

# 기본 확인
print(data.head())
print(data.info())
```
